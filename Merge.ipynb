```python
import pandas as pd 
```

```python
df1 = pd.read_csv('Ted.csv')
df2 = pd.read_csv('TedX.csv')
```

```python
def num_conv(lst, name):

    l = []
    for i in lst:
        if(i[-1] == 'K'):
            l.append(int(float(i[:-1]) * 1000))
        elif(i[-1] == 'M'):
            l.append(int(float(i[:-1]) * 1000000))
        else:
            l.append(int(i))  
            
    return pd.DataFrame(l, columns = [name])
```

```python
title  = df1['title']
link   = df1['link']
author = df2['author']
date   = df2['date']

likes = df2['likes']
likes = num_conv(likes, 'likes')
views  = df2['views']
views = num_conv(views, 'views')
```

```python
df = pd.concat([title, author, date, views, likes, link], axis = 1)
```
